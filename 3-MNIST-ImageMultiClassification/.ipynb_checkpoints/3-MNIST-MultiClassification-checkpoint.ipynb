```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits # mnist
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost
from xgboost  import XGBClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,make_scorer, f1_score, roc_auc_score

import warnings
warnings.filterwarnings("ignore")
```

### Tarin Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print("X_train: {}".format(X_train.shape))
print("X_test: {}".format(X_test.shape))
print("y_train: {}".format(y_train.shape))
print("y_test: {}".format(y_test.shape))
```

### Data Standardization

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Dimension reduction with PCA (Principal Component Analysis) - Boyut İndirgeme

 **64 featuren -> n feature** düşürmeye çalışıcaz.
PCA (Principal Component Analysis),  yüksek boyutlu veriyi daha düşük boyutlara indirerek veri kaybını en aza indirmeyi amaçlayan bir boyut indirgeme yöntemidir.

Nasıl Çalışır?
1.  Veriyi Standartlaştırır → Özellikleri aynı ölçeğe getirir.
2.	Kovaryans Matrisini Hesaplar → Özellikler arasındaki ilişkileri inceler.
3.	Özdeğer ve Özvektörleri Bulur → En fazla varyansı açıklayan yönleri belirler.
4.	Ana Bileşenleri Seçer → En önemli bileşenleri seçerek boyutu azaltır.


```python
pca = PCA(n_components=0.95)
```

- Veri setinin toplam varyansının %95’ini koruyacak şekilde yeterli sayıda ana bileşen (principal component) seçilir.
- Elle belirli bir bileşen sayısı seçmek yerine modelin gerekli sayıda bileşen seçmesine izin veririz.

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
print("X_train_scaled: {}".format(X_train_scaled.shape))
print("X_train_pca: {}".format(X_train_pca.shape))
print("X_test_scaled: {}".format(X_test_scaled.shape))
print("X_test_pca: {}".format(X_test_pca.shape))
```

**64 feature** → **40 feature** (PCA ile boyut indirgeme)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_train_pca[:, 0], y=X_train_pca[:, 1], hue=y_train, palette="viridis", alpha=0.7)
plt.xlabel("PCA 1. Bileşen")
plt.ylabel("PCA 2. Bileşen")
plt.title("PCA Sonrası Verinin Sınıflara Göre Dağılımı")
plt.legend(title="Sınıflar")
plt.show()
```

## Model  Trainining And GridSearch

### Modelling

```python
models = [("LOJ",LogisticRegression()),
          ("KNN",KNeighborsClassifier()),
          ("SVC",SVC(probability=True)),
          ("MLP",MLPClassifier()),
          ("GBM",GradientBoostingClassifier()),
          ("XGB",XGBClassifier()),
          ("CATB",CatBoostClassifier(verbose=0))
          ]
```

```python
finalResult = pd.DataFrame(columns=["Model","Accuracy"]) # score
# cmList = [] # Confusion matrix

for name,model in models:
    model.fit(X_train_pca,y_train) # training
    y_pred = model.predict(X_test_pca)
    acc = accuracy_score(y_test,y_pred)
    finalResult = pd.concat([finalResult,pd.DataFrame([[name,acc*100]],columns=["Model","Accuracy"])],
                               ignore_index=True)
    #cm = confusion_matrix(y_test,y_pred)
    #cmList.append((name,cm))
```

```python
finalResult
```

### Hyperparameter Tuning - GridSearchCV

```python
model_params = {
     "SVM":{
        "model":SVC(probability=True),
        "params":{ # 3x2=6
            "C":[0.1, 1, 10],
             "kernel":["linear","rbf"] 
        }
    },
    "KNN":{
        "model":KNeighborsClassifier(),
        "params":{# 6
            "n_neighbors":[1,3,5,7,9,11]# odd numbers
        
        }
    },
     "Loj":{
        "model":LogisticRegression(max_iter=2000),
        "params":{
            "penalty":["l2"], # default='l2'
            "C":[0.1, 1, 10, 100], # default=1.0
            "solver": ["lbfgs","saga"], #default='lbfgs'
        } # 4x2=8
    },
     "MLP":{
        "model":MLPClassifier(max_iter=2000),
        "params":{
            "alpha":[0.001, 0.01, 0.1, 1],
            "hidden_layer_sizes":[(10,10),(100,100),(128,128,128)]  
        }# 4x3=12
    },
    "CatBoost":{
        "model":CatBoostClassifier(),
        "params":{
            "iterations":[500, 1000],
            "learning_rate":[0.01,0.1],
            "depth":[4,6,8]
        }# 2x2x3=12
    },

    "XGB":{
        "model":XGBClassifier(),
        "params":{
            "n_estimators":[100,500,1000],
            "subsample":[0.8,1],
            "max_depth":[3,5,7],
            "learning_rate":[0.1,0.01]
        }# 3x2x3x2=36
    },
   "GBM":{
        "model":GradientBoostingClassifier(),
        "params":{
            "learning_rate":[0.1,0.01,0.05],
            "n_estimators":[100,500,1000],
            "max_depth":[5,8],
            "subsample":[0.6,1.0]
        }# 3x3x2x2=36
    }, 
}
```

- SVM: 6x5=30
- KNN: 6×5=30
- Logistic Regression: 8×5=40
- MLP: 12×5=60
- CatBoost: 12×5=60
- XGB: 36×5=180
- GBM: 36x5=180

Total Fit Count => 30+30+40+60+60+180+180=580

* **F1-micro**, tüm sınıfların toplam TP, FP ve FN değerleri üzerinden genel bir ortalama hesaplar.

```python
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'f1_micro': make_scorer(f1_score, average='micro'), 
   # 'roc_auc': make_scorer(roc_auc_score, multi_class='ovo', needs_threshold=True)  
}

columns = ["model", "best_params", "accuracy", "f1_micro"]
scores = []

for model_name, params in model_params.items():
    print(f"Training {model_name}...")
    grid_search = GridSearchCV(estimator=params["model"], 
                               param_grid=params["params"], 
                               scoring=scoring,
                               refit="accuracy", #best model metric
                               cv=5, n_jobs=-1)
    
    grid_search.fit(X_train_pca, y_train)

    best_index = grid_search.best_index_
    f1_micro = grid_search.cv_results_.get("mean_test_f1_micro", [None])[best_index]
  #  roc_auc = grid_search.cv_results_.get("mean_test_roc_auc", [None])[best_index]

    scores.append([model_name, dict(grid_search.best_params_), grid_search.best_score_,f1_micro])

df_scores = pd.DataFrame(scores, columns=columns)
```

## Voting Classifier 

```python
voting_models = [
    ('svm', SVC(C=10, kernel="rbf", probability=True)),
    ('mlp', MLPClassifier(alpha=0.1, hidden_layer_sizes=(100,100))),
    ('knn', KNeighborsClassifier(n_neighbors=3)),
    ('catboost', CatBoostClassifier(depth=8, iterations=1000, learning_rate=0.1, verbose=0)),
    ('logreg', LogisticRegression(C=1, penalty='l2', solver='saga')),
    ('xgb', XGBClassifier(learning_rate=0.1, max_depth=3, n_estimators=500, subsample=0.8)),
    ('gbm', GradientBoostingClassifier(learning_rate=0.05, max_depth=5, n_estimators=500, subsample=0.6))
]
```

```python
soft_voting_clf = VotingClassifier(estimators=voting_models, voting='soft')

soft_voting_clf.fit(X_train_pca, y_train)

y_pred = soft_voting_clf.predict(X_test_pca)
```

- hard : çoğunluk oyuna göre 
- soft : ortalama, olasılık oylama
