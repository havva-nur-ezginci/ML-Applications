## Outlier Detection (Aykırı Değer Tespiti) : Z-skor,Winsorizing

```python
sns.boxplot(data= df.loc[:, numeric_list])
plt.show()
```

```python
numeric_list
```

**Z-skor**
- Z-skoru, bir veri noktasının ortalamadan kaç standart sapma uzaklıkta olduğunu gösterir. Z-skoru genellikle 3'ten büyük olan değerler aykırı değer olarak kabul edilir.

```python
# outlier detection
z_scores = (df[numeric_list] - df[numeric_list].mean()) / df[numeric_list].std()
outliers = (z_scores.abs() > 3)
outlier_rows = df[outliers.any(axis=1)]

print("Outlier Değerler:")
print(outlier_rows[numeric_list])
```

**Winsorizing**
- Veri sayısının yetersiz olduğundan aykırı gözlemlerin kaldırılması yerine düzeltilmesi tercih edilmiştir. Verinin düzeltilmesi için
ise Winsorizing yöntemi kullanılmıştır.
- Winsorizing tekniği, aykırı değerleri bir sınırla sınırlamak anlamına gelir. Aykırı değerler, belirli bir üst ve alt yüzdeyi temsil eden sınırlarla değiştirilir.

```python
# Winsorizing 
for column in numeric_list:

    lower_limit = df[column].quantile(0.05)  # bottom %5
    upper_limit = df[column].quantile(0.95)  # Top %5
    
    df[column] = np.where(df[column] < lower_limit, lower_limit, df[column])
    df[column] = np.where(df[column] > upper_limit, upper_limit, df[column])
```

```python
z_scores = (df[numeric_list] - df[numeric_list].mean()) / df[numeric_list].std()
outliers = (z_scores.abs() > 3)
outlier_rows = df[outliers.any(axis=1)]

print("Outlier Değerler:")
print(outlier_rows[numeric_list])
```

### Encoding Categorical Columns

### Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
print("X_train: {}".format(X_train.shape))
print("X_test: {}".format(X_test.shape))
print("y_train: {}".format(y_train.shape))
print("y_test: {}".format(y_test.shape))
```

## Modelling

```python
models = [("DIC",DecisionTreeClassifier()),
          ("RFC",RandomForestClassifier()),
          ("LOJ",LogisticRegression()),
          ("KNN",KNeighborsClassifier()),
          ("SVC",SVC()),
          ("MLP",MLPClassifier()),
          ("CART",DecisionTreeClassifier()),
          ("GBM",GradientBoostingClassifier()),
          ("XGB",XGBClassifier()),
          ("LGBM",LGBMClassifier()),
          ("CATB",CatBoostClassifier())
          ]
```

```python
finalResult = pd.DataFrame(columns=["Model","Accuracy","Precision","Recall"]) # score
cmList = [] # Confusion matrix

for name,model in models:
    if(name == "MLP" or name=="KNN" or name=="SVC" or name=="LOJ"):
        # X_train -  X_test scaled
        scaler = StandardScaler()
        X_train_scaled = X_train
        X_test_scaled = X_test
        # sadece nümerik değerleri scale edicez
        X_train_scaled[numeric_list[:-1]] = scaler.fit_transform(X_train_scaled[numeric_list[:-1]])
        X_test_scaled[numeric_list[:-1]] = scaler.fit_transform(X_test_scaled[numeric_list[:-1]])

        model.fit(X_train_scaled,y_train) # training
        y_pred = model.predict(X_test_scaled) # prediction
        
        acc = accuracy_score(y_test,y_pred)
        score = precision_score(y_test,y_pred)
        recall = recall_score(y_test, y_pred)
        finalResult =pd.concat([finalResult,pd.DataFrame([[name,acc*100,score,recall]],columns=["Model","Accuracy","Precision","Recall"])],
                               ignore_index=True)

        cm = confusion_matrix(y_test,y_pred)
        cmList.append((name,cm))
    else:

        model.fit(X_train,y_train) # training
        y_pred = model.predict(X_test) # prediction
        
        acc = accuracy_score(y_test,y_pred)
        score = precision_score(y_test,y_pred)
        recall = recall_score(y_test, y_pred)
        finalResult =pd.concat([finalResult,pd.DataFrame([[name,acc*100,score,recall]],columns=["Model","Accuracy","Precision","Recall"])],
                               ignore_index=True)

        cm = confusion_matrix(y_test,y_pred)
        cmList.append((name,cm))
```
