## Veriyi incelemek

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram,linkage
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['CustomerID'], inplace=True)
```

```python
df.shape
```

```python
df.describe().T
```

### Encoding Categorical Column

```python
# Binary Encoding (female → 0, male → 1)
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
df.head()
```

### Visualising Data

```python
df.hist(figsize=(10,10))
```

```python
# Gelir & Harcama Skoru İlişkisi
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Annual Income (k$)'], 
                y=df['Spending Score (1-100)'], 
                hue=df['Gender'], palette='coolwarm', s=100)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Annual Income & Spending Score")
plt.show()
```

```python
df['SpendingCategory'] = pd.cut(df['Spending Score (1-100)'], bins=[0, 40, 70, 100], labels=['Low', 'Medium', 'High'])
df.groupby('SpendingCategory').mean()
```

* 0 - 40 → "Low" (Düşük Harcama)
* 41 - 70 → "Medium" (Orta Harcama)
* 71 - 100 → "High" (Yüksek Harcama)

* Low :
  - Müşterilerin %46’sı erkek, %54’ü kadın.
  - Geliri yüksek olmasına rağmen harcamayanlar.


* Medium :
  - Müşterilerin %44.6’sı erkek, %55.4’ü kadın.

* High :
  - Müşterilerin %40.7’si erkek, %59.3’ü kadın.
  - Geliri yüksek olan genç müşteriler en fazla harcamayı yapıyor.


**Kadınların harcama skorları daha yüksek. "High" harcama kategorisinde kadın oranı daha fazla.**

**Daha genç müşteriler daha fazla harcama yapıyor.**

**Gelir yüksek ama yaşlı müşteriler harcamıyor, gençler ise daha fazla harcıyor.**

```python
sns.countplot(x="SpendingCategory", hue="Gender", data=df, palette="coolwarm")

plt.title("Customer Distribution by Spending Category and Gender")
plt.xlabel("Spending Category")
plt.ylabel("Number of Customers")
plt.legend(["Female", "Male"])

plt.show()
```

### Feature Scaling (Standardizasyon)

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.iloc[:,0:4])
```

```python
df_scaled[0:10]
```

### Temel Bilesen Analizi (Principle Component Analysis (PCA))

```python
pca = PCA().fit(df_scaled)
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', color='b', linestyle='--')

plt.title('Kümülatif Varyans Oranı (Cumulative Explained Variance)', fontsize=14)
plt.xlabel("Principal Components", fontsize=12)
plt.ylabel("Cumulative Explained Variance Ratio", fontsize=12)
plt.grid(True)
plt.show()
```

**Bileşen sayısı arttıkça, veri kümesi içindeki varyansın (varyans: veri kümesinin sahip olduğu bilgi) artan sayıda bileşen tarafından açıklandığını ve sonunda tüm veri kümesini kapsadığını gözlemleyebiliriz.**

```python
pca.explained_variance_ratio_
```

```python
# final model
pca = PCA(n_components=4)
pca_fit = pca.fit_transform(df_scaled)
pca.explained_variance_ratio_
```

* 3 bileşen ≈ %83.18 (bir miktar kayıp)
* 4 bileşen = %100 

## Determining the Optimal Number of Clusters : Yellowbrick 

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k=(2,10))  
visu.fit(df_scaled)
visu.poof()
```

## Clustring Modeling : Unspervised learning (Kmeans)

```python
df_scaled[0:5]
```

```python
df_scaled.shape
```

```python
kmeans = KMeans(n_clusters=5)
df['Cluster'] = kmeans.fit_predict(df_scaled)
cluster_centers = kmeans.cluster_centers_
print(df["Cluster"])
```

```python
cluster_centers
```

```python
kmeans.get_params()
```

```python
df.head()
```

```python
kmeans.labels_
```

```python
#  Visualisation with 2D PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df['Cluster'], palette='viridis', s=100, edgecolor='black')
plt.scatter(pca.transform(cluster_centers)[:,0], pca.transform(cluster_centers)[:,1], 
            s=300, c='red', marker='X', label="Cluster Centers")

plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Customer Segmentation with K-Means (PCA Reduced)")
plt.legend()
plt.show()
```

## Hiyerarşik Kümeleme : Dendrogram

```python
hc_complete = linkage(df_scaled,"complete")

plt.figure(figsize=(10,8)) 
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Veri Noktaları")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete,
           truncate_mode= "lastp", # en son p adet küme göster
           p = 10, 
           show_contracted= True,
          leaf_font_size=10,
          ); 
plt.grid(True) 
plt.show()
```

Kümeler arasındaki uzaklık farkı en fazla olan yatay çizgiyi (kırılım noktası) belirleyerek optimum küme sayısını seçilmelidir.

**Bu dendrogramda => 4 veya 5 küme**
